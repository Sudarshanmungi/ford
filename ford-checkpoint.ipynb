```python
# Count occurrences of each fuel type
fuel_type_counts = df['fuelType'].value_counts()

# Plot bar graph for distribution of fuel types
plt.figure(figsize=(10, 6))
sns.barplot(x=fuel_type_counts.index, y=fuel_type_counts.values, palette='pastel')
plt.title('Distribution of Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot for price vs mileage
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='mileage', y='price', hue='fuelType', palette='Set2')
plt.title('Price vs Mileage')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.legend(title='Fuel Type')
plt.show()
```

```python
# Calculate total mileage for each transmission type
transmission_mileage = df.groupby('transmission')['mileage'].sum()

# Plot pie chart for transmission type distribution based on mileage
plt.figure(figsize=(8, 8))
plt.pie(transmission_mileage, labels=transmission_mileage.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), explode=(0.05, 0.05, 0.05))
plt.title('Distribution of Transmission Types based on Mileage')
plt.show()
```

```python
# Hypothetical data: Average price over the years
average_price_by_year = df.groupby('year')['price'].mean()

# Convert index to numeric for proper sorting
average_price_by_year.index = pd.to_numeric(average_price_by_year.index)

# Sort the data by year
average_price_by_year = average_price_by_year.sort_index()

# Create line plot
plt.figure(figsize=(8, 4))
sns.lineplot(x=average_price_by_year.index, y=average_price_by_year.values, marker='o', color='skyblue', linestyle='-')
plt.title('Average Price of Vehicles Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Remove non-numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Create a heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
